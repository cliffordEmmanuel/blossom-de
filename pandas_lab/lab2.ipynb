```python
# pd.join()

left = pd.DataFrame(
    {"A": ["A0", "A1", "A2"], "B": ["B0", "B1", "B2"]}, index=["K0", "K1", "K2"]
)
left
```

```python
right = pd.DataFrame(
    {"C": ["C0", "C2", "C3"], "D": ["D0", "D2", "D3"]}, index=["K0", "K2", "K3"]
)
right
```

```python
left.join(right)
```

```python
left.join(right, how='outer')
```

```python
left.join(right, how='inner')
```

## grouping

```python
us_companies.head()
```

```python
# count of all the companies per city?

company_per_city = us_companies.groupby(['state','city'], as_index=False).agg({'company_name_id':'count'}) # .to_csv('output/company_percity.csv')
company_per_city
```

```python
us_companies.columns
```

```python
l = [[1, 2000, 3], [1, 8000, 40], [2, 30, 30], [2, 3000, 27]]
df = pd.DataFrame(l, columns=["customerid", "amount", "age"])
df
```

```python
pd.pivot_table(
    df,
    index=['customerid'],
    values=['amount'],
    aggfunc={'amount':'sum','customerid':'count'},
)
```

```python
pd.pivot_table(
    us_companies,
    index=['state','city'],
    values=['company_name_id'],
    aggfunc={'company_name_id':'count'},
).reset_index()
```

```python
company_per_city.sort_values(by=['state','city','company_name_id'], ascending=True)
```
